```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('Salary_Data.csv')
df.head()
```

```python
plt.scatter(df["YearsExperience"], df['Salary'])
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.title('YearsExperience vs Salary')
plt.show()
```

```python
X = df[['YearsExperience']]
y = df['Salary']
```

```python
X,y
```

```python
from sklearn.model_selection import train_test_split#converts into test ing and traing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state = 0)#x in i and y is o,o.30 m4asn 70 percent for train and 30 for testing random state ids random shuffle.random state must be integer
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model
```

```python
model.fit(X_train, y_train)
```

```python
print("The slope or coefficient of yrs of exp is: ", model.coef_)
print("The intercept of the line is: ", model.intercept_)
```

```python
plt.scatter(X_train, y_train)
plt.plot(X_train, model.predict(X_train), 'g')#g green color
plt.show()
```

 Why does X_train and X_test come from the SAME dataframe?

Because

 You split your original dataset into two parts:

X_train  data the model will learn from (SEEN)

X_test  data the model will be tested on (UNSEEN)

But both parts come from the same original dataset you collected.

"Unseen" does NOT mean "new data from outside the dataset".

It means:

 Your model has NOT seen these rows while training

(not that the rows come from some magical new dataset) Unseen  From different dataset
 Unseen = Not used during training

```python
plt.scatter(X_test,y_test)
plt.plot(X_test,model.predict(X_test),"r")
plt.show()
```

```python
y_pred_test = model.predict(X_test)
```

```python
y_pred_test,y_test
```

```python
plt.scatter(X_test, y_test)#blue wala
plt.scatter(X_test, model.predict(X_test), c= 'r')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
mse = mean_squared_error(y_test, y_pred_test)
mae = mean_absolute_error(y_test, y_pred_test)
rmse = np.sqrt(mse)
print("Mean Squared Error: ", mse)
print("Mean Absolute Error: ", mae)
print("Root Mean Squared Error: ", rmse)
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test, y_pred_test)
print(score)
print("R2 Score: ", score * 100, '%')
```

```python
# Display Adjusted R2 Score
adjusted_r2 = 1 - (1 - score) * (len(y_test) - 1) / (len(y_test) - X_test.shape[1] - 1)
print("Adjusted R2 Score: ", adjusted_r2 * 100, '%')
```
